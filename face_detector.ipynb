```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('data/images/face.afa6543c-2933-11ec-bfe6-30c9ab8f951a.jpg')
```

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces = face_detector.detectMultiScale(img, 1.1, 6)
```

```python
faces
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(np.squeeze(img))
```

```python
def draw_rect(img, faces, sf=20):
    for x, y, w, h in faces:
        x = x - sf
        y = y - 2*sf
        cv2.rectangle(img, (x,y), (x+w+2*sf, y+h+3*sf), (255, 0 , 0), 3)
```

```python
draw_rect(img, faces)
```

```python
plt.imshow(np.squeeze(img))
```

```python
faces
```

```python
def save_crop(img, faces, sf=20):
    i=0
    for (x, y, w, h) in faces:
        x = x - sf
        y = y - 2*sf
        img2 = np.squeeze(img)
        crop = img2[y:y+h+3*sf, x:x+w+2*sf]
        crop = cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)
#         plt.imshow(np.squeeze(crop))
        cv2.imwrite('crop_img/crop'+str(i)+'.png',crop)
        i+=1
```

```python
save_crop(img, faces)
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    _, frame = cap.read()


    faces = face_detector.detectMultiScale(frame, 1.1, 4)
    print(faces)

    draw_rect(frame, faces)

    # Display the output
    cv2.imshow('Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
```
